Write a function to get a sample from a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution_sample():
    data = np.random.normal(0, 1)
    plt.hist(data)
    plt.show()

normal_distribution_sample()

```

```python
import scipy.stats
import matplotlib.pyplot as plt

def gen_nplot(N):
  dist = scipy.stats.norm(0,1)
  data = dist.rvs(N)
  plt.hist(data,bins=int(N/10))
  plt.xlabel("Random number")
  plt.ylabel("Frequency")
  plt.show()

gen_nplot(100000)
```
